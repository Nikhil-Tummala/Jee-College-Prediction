```python
def clean_rank(value):
    try:
        return int(float(value))
    except ValueError:
        return int(value[:-1]) if value[:-1].isdigit() else np.nan
```

```python
final_df['Opening Rank'] = final_df['Opening Rank'].apply(clean_rank)
```

```python
final_df['Closing Rank'] = final_df['Closing Rank'].apply(clean_rank)
```

```python
final_df["Closing Rank"].isnull().sum()
```

```python
X = final_df[['Opening Rank', 'Gender', 'Seat Type']]
y = final_df[['Institute', 'round']]


categorical_features = ['Gender', 'Seat Type']
numeric_features = ['Opening Rank']
```

```python
le_institute = LabelEncoder()
final_df['Institute'] = le_institute.fit_transform(final_df['Institute'])
```

```python

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])
```

```python
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', MultiOutputClassifier(RandomForestClassifier(n_estimators=50, random_state=42)))
])
```

```python
train_data = final_df[final_df['year'] < final_df['year'].max()]
test_data = final_df[final_df['year'] == final_df['year'].max()]
```

```python
X_train = train_data[['Opening Rank', 'Gender', 'Seat Type']]
y_train = train_data[['Institute', 'round']]
X_test = test_data[['Opening Rank', 'Gender', 'Seat Type']]
y_test = test_data[['Institute', 'round']]
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)
```

```python
from joblib import dump, load

# Save the trained model to a file using joblib
dump(model, 'jee_model.joblib')
```

```python
from sklearn.metrics import accuracy_score, classification_report

for target in y_test.columns:
    print(f"Results for {target}:")
    print(classification_report(y_test[target], y_pred[target]))
```

```python
def get_top_n_predictions(model, X, n=5):
    probas = model.predict_proba(X)
    top_n_predictions = []

    for i in range(X.shape[0]):
        preds = {}
        for j, target in enumerate(y_train.columns):
            top_n_indices = np.argsort(probas[j][i])[::-1][:n]
            top_n_preds = [model.classes_[j][index] for index in top_n_indices]
            preds[target] = top_n_preds
        top_n_predictions.append(preds)
    
    return top_n_predictions
```

```python
top_n_predictions = get_top_n_predictions(model, X_test, n=5)

# Print top-5 predictions for the first test sample
print(top_n_predictions[0])

# Decode the 'Institute' predictions back to the original labels
decoded_predictions = [le_institute.inverse_transform(pred['Institute']) for pred in top_n_predictions]

# Print the decoded predictions for the first test sample
print(decoded_predictions[0])
```
